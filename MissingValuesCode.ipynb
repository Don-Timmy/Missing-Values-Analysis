# Percentage of Missing Values

```python
missinglocustdata = pd.DataFrame(locustdata.isnull().sum(), columns = ['Number of Missing Observations'])
missinglocustdata['Number of Taken Observations'] = (locustdata.shape[0] - missinglocustdata['Number of Missing Observations'])
missinglocustdata['percent'] = round((missinglocustdata['Number of Missing Observations']/locustdata.shape[0]) * 100 , ndigits = 0)
missinglocustdata.sort_values(ascending = False, by = 'percent', inplace = True)
missinglocustdata
```

## Percentage of Missing Values Filter

```python
missinglocustdata[missinglocustdata['percent'] > 50]
```
